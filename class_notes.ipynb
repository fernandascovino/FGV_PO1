#### Algoritmo de varredura

$$x^{t+1}= x^{t}+ \lambda d$$
Num espaço contínuo, a direção ($d$) mais promissora é o gradiente, mas que acaba dependendo do tamanho do passo ($\lambda$). Também, não necessariamente o gradiente está dentro da região factível.

* Região factível: conjunto de soluções que não viola nenhuma restrição

#### Simplex (pg. 65 - Gentle): 
$$x^{t+1}= x^{t}+ \lambda d$$

* Tamanho do passo: partindo de um ponto inicial, um passo Simplex é a primeira interseção com alguma das restrições na direção escolhida.


* Direção promissora: comparamos o ângulo da direção com o gradiente naquele ponto. Se o ângulo estiver em -/2, /2.


* Exemplo $max\{ z(x) = c^t x + w: Ax = b, x \geq 0\}$ **(sempre na forma padrão! w uma cte)**:

Antes de provar, vamos ver um exemplo:

4 x1 + 10 x2 - 6x3 - 2x4 = 6
-2x1 + 2x2 - 4x3 + x4 = 5
-7x1 - 2x2 + 4x4 = 3

x1, x2, x3, x4  0

Afirmamos: esse problema é infactível. Mas como podemos ver isso?

Tomemos o vetor de coeficientes v = [1, -2, 1]. Fazendo a multiplicação com as equações, temos:

(-1) * 4 x1 + 10 x2 - 6x3 - 2x4 = (-1) * 6 
(-2) * -2x1 + 2x2 - 4x3 + x4 = (-2) * 5   +
(1) * -7x1 - 2x2 + 4x4 = (1) * 3
___________________________
x1 + 4x2+ 2x3 = -1

Vemos que chegamos a uma equação infactível. Logo, podemos provar que o problema é factível achando esse conjunto de números do vetor v.

Teoricamente eficiente, mas a implementaçãoé ineficaz.

Branch and Bound

Pegamos a cota superior (BU)  e inferior (BL) da solução dada para uma das variáveis (xi), e criamos novas restrições, removendo essas regiões não inteiras:

xi ≤ BL
xi ≥ BU

E fazemos isso nas variáveis, criando uma árvore de decisão, até achar uma solução inteira ótima (cuja função objetivo seja maximizada).

Programação inteira

A principal diferença é que as variáveis não são contínuas, o que muda nosso desenho e abordagem do problema. Em geral, resolvemos  fazendo uma combinação de problemas lineares.

big M 

Quando queremos condicionar a existência de uma variável positiva à outra (SE), utilizamos a técnica do big M: 

Xj ≤M × Xi

O coeficiente  M pode ser um positivo qualquer, que restringe Xj em relação a Xi.

Múltipla escolha

Exemplo:
máx X1 + X2 + X3 + X4 
sujeito a  X1 + X2 + X3 + X4 ≤ 1

Podemos usar essa técnica para escolher restrições (OU) também, colocando variáveis binárias auxiliares:
máx X1 + X2 + X3 + X4 
sujeito a  Y1 + Y2 + Y3 = 1
Yi ∈ {0,1}
rest1 < b1 × Y1
rest2 < b2 × Y2
rest3 < b3 × Y3
